```
data['Trend']=range(151)
for i in range(data.shape[0]):
  a=data['Day_Perc_Change'][i]
  if -0.5<a<0.5:
    data['Trend'][i]='Slight or No change'
  elif 0.5<a<1:
    data['Trend'][i]='Slight positive'
  elif -1<a<-0.5:
    data['Trend'][i]='Slight Negative'
  elif 1<a<3:
    data['Trend'][i]='Positive'
  elif -3<a<-1:
    data['Trend'][i]='Negative'
  elif 3<a<7:
    data['Trend'][i]='Among top gainers'
  elif -7<a<-3:
    data['Trend'][i]='Among top losers'
  elif a>7:
    data['Trend'][i]='Bull run'
  elif a<-7:
    data['Trend'][i]='Bear drop'
                               
```

```
data.head()
```

```
def total_traded_quantity(data):
  mean,median=[],[]
  a=data['Trend'].unique()
  for i in a:
    new=data.loc[data['Trend']==i,'Total Traded Quantity']
    mean.append(np.mean(new))
    median.append(np.median(new))
  result=pd.DataFrame({'Trend':a,'mean':mean,'median':median})
  return result  
```

```
result=total_traded_quantity(data)
print(result)
```
